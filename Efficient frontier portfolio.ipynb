```python
results = []

for _ in range(5000):
    weights = np.random.random(len(mean_returns))
    weights /= np.sum(weights)

    ret = np.dot(weights, mean_returns)
    vol = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe = ret / vol

    results.append((ret, vol, sharpe, weights))
```

```python
rets = [r[0] for r in results]
vols = [r[1] for r in results]
sharpes = [r[2] for r in results]
```

```python
import matplotlib.pyplot as plt

plt.scatter(vols, rets, c=sharpes)
plt.xlabel("Risk")
plt.ylabel("Return")
plt.colorbar(label="Sharpe")
plt.show()
```

```python
max_sharpe_port = max(results, key=lambda x: x[2])
```

```python
from scipy.optimize import minimize
import numpy as np

def portfolio_performance(weights):
    ret = np.dot(weights, mean_returns)
    vol = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    return ret, vol


def max_sharpe_constrained():
    n = len(mean_returns)

    def neg_sharpe(weights):
        ret, vol = portfolio_performance(weights)
        return -(ret / vol)

    # weights must sum to 1
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

    # 🔥 max 10% per stock
    bounds = tuple((0, 0.1) for _ in range(n))

    # starting point
    init_guess = n * [1./n]

    result = minimize(neg_sharpe, init_guess,
                      method='SLSQP',
                      bounds=bounds,
                      constraints=constraints)

    return result.x
```

```python
weights_constrained = max_sharpe_constrained()
```

```python
for stock, w in zip(data.columns, weights_constrained):
    print(stock, round(w*100, 2), "%")
```

```python
n = len(mean_returns)

weights_equal = np.ones(n) / n

ret_eq = np.dot(weights_equal, mean_returns)
vol_eq = np.sqrt(np.dot(weights_equal.T, np.dot(cov_matrix, weights_equal)))
sharpe_eq = ret_eq / vol_eq
```

```python
# Unconstrained
ret_u, vol_u = portfolio_performance(max_sharpe_port[3])
sharpe_u = ret_u / vol_u

# Constrained
ret_c, vol_c = portfolio_performance(weights_constrained)
sharpe_c = ret_c / vol_c

# Equal already computed

print("----- Comparison -----")

print("Unconstrained:")
print("Return:", ret_u, "Vol:", vol_u, "Sharpe:", sharpe_u)

print("\nConstrained (10% cap):")
print("Return:", ret_c, "Vol:", vol_c, "Sharpe:", sharpe_c)

print("\nEqual Weight:")
print("Return:", ret_eq, "Vol:", vol_eq, "Sharpe:", sharpe_eq)
```
